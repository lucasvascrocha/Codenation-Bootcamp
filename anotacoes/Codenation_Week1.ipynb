# Big Data - O Grande problema

Os 5V'S da big data:

* Velocidade -  velocidade que os dados estão sendo criados 
* Volume - o volume dos dados que aumentam a cada segundo
* Variety - Variedade dos dados, dados estruturados e sem estruturura
* Veracity - O quão esses dados são confiaveis
* Value - Valor que esses dados possuem

# Tipos de Solução de Análise


* Solução Descritiva (Descriptive Analytics)

  * Descreve o que existe;
  * Olha apenas para o passado/atual;
  * " O que aconteceu ? "


* Solução Diagnóstica

  * Descreve os porquês;
  * Olha apenas para o passado/atual;
  * "Porque isso aconteceu?"


* Solução Preditiva ( Predictive Analytics )

  * Descreve o que vai existir;
  * Usa o passado para prever/predizer o futuro;
  * "O que vai acontecer?"

* Soluções Prescritivas ( Prescriptive Analytics )

  * Recomenda tomada de decisões;
  * Usando uma predição/previsão recomenda uma ação
  * "O que vou fazer?"

Material: 4 Types of Data Analytics


# Tipos de Problemas?

* Classificação binária

  * Preveem um resultado binário ( Uma de duas classes possíveis )

  * Regressão Logística

* Classificação multiclasse

  * Permite gerar previsões para várias classes ( Prever um entre mais de dois resultados);

  * Regressão Logística Multinomial.


* Regressão

  * Modelo de regressão
 
  * Preveem um valor númerico

  * Regressão linear.

* Agrupamento / Clusterização

  * Modelo de aprender grupos / divisões nos dados

  * Retorna um grupo

  * KMeans

* Sistemas de Recomendação
  * Modelo sugere/recomenda algo à um usuário

  * Retorna sugestão

  * Collaborative filtering / Content based


# Tipos de Aprendizado

* Aprendizado Supervisionado: São apresentadas ao computador exemplos de entradas e saídas desejadas, fornecidas por um "professor". O objetivo é aprender uma regra geral que mapeia as entradas para as saídas.

* Aprendizado Não - Supervisionado: Nenhum tipo de etiqueta é dado. Encontra estrutura nas entradas fornecidas sozinho. Pode ser um objetivo em si mesmo (descobrir novos padrões nos dados) ou um meio para atingir um fim.


# Maturidade com Analytics

* Raw Data = Os dados da empresa geram pouco valor.

* Informação = Dados começam a trazer algum valor pra empresa mas os relatorios não são automatizados.

* Bussiness Intelligence  = Uma visão limpa do passado, presente e possuem governaça de dados

* Advanced Analytics - Possuem conhecimento profundos sobre os dados e analises preditivas.
